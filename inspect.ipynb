```python
import pandas as pd
```

```python
file_path = 'data/order/am_hq_order_spot.txt'
column_name = ['col1', 'col2', 'col3', 'col4', 'col5', 'col6', 'col7', 'ApplSeqNum', 
               'SecurityID', 'col8', 'Price', 'OrderQty', 'TransactTime', 'Side', 'OrderType',
               'col9', 'col10', 'col11', 'col12', 'col13']
# Set all column dtype as string
df = pd.read_csv(file_path, sep='\t', header=None, names=column_name, dtype=str)
df.head()
```

```python
# Select Valid data
df = df[df['SecurityID'] == '000001']
df = df[(df['TransactTime'] >= '20190102093000000') & (df['TransactTime'] < '20190102113100000')]
df[df['OrderType'] == '1']['Price'].value_counts()
```
